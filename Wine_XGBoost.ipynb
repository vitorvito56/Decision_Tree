```python
#This is a classification tree, using the XGBoost to help it, using the wine dataset

import pandas as pd
```

```python
import sklearn.datasets as datasets

df = datasets.load_wine()
data = pd.DataFrame(data=df.data, columns=df.feature_names)
data

#Just importing the dataset
```

```python
data.dropna()
data

#Droping any missing values (what it didnt had)
```

```python
Target = df.target
Target

#Choosing the Target, what we are gonna predict
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(data,Target,test_size=0.3, random_state=35)

#Separating the training and test, if 30% of the data being for test.
```

```python
#pip install xgboost

#I needed to install before using
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(x_train, y_train)

#Applyng the XGBoost
```

```python
accuracy = model.score(x_test, y_test)
print(f"Accuracy= {accuracy}")

#Checking and seing a pretty good accuracy
```

```python
help(XGBClassifier())

#Checking what I can change to see how the model works
```

```python
#Now im repeating the process, but with a subsample to prevent overfitting, with this i good a PERFECT model.

x_train,x_test,y_train,y_test= train_test_split(data,Target,test_size=0.3, random_state=35)
model = XGBClassifier(subsample=0.6)
model.fit(x_train, y_train)
accuracy = model.score(x_test, y_test)
print(f"Accuracy= {accuracy}")
```

```python
#Checking with max subsample, and we had the first inital result.

x_train,x_test,y_train,y_test= train_test_split(data,Target,test_size=0.3, random_state=35)
model = XGBClassifier(subsample=1)
model.fit(x_train, y_train)
accuracy = model.score(x_test, y_test)
print(f"Accuracy= {accuracy}")
```

```python
#Using know the reg_alpha and the reg_lambda, that punishes the model in the loss function.
#When they values are low it changes nothing, but with grat values, we have a decrease on accuracy

x_train,x_test,y_train,y_test= train_test_split(data,Target,test_size=0.3, random_state=35)
model = XGBClassifier(
    reg_alpha=3,     # L1 regularization (Lasso)
    reg_lambda=2,    # L2 regularization (Ridge)
)
model.fit(x_train, y_train)
accuracy = model.score(x_test, y_test)
print(f"Accuracy= {accuracy}")

```
